# Order Prioirity


```python
df['Category'].value_counts()
```

```python
plt.figure(figsize=(8, 8))  
df7 = sns.countplot(x='Category', data=df)
for bars in df7.containers:
    df7.bar_label(bars)
plt.show()
```

```python
size_distribution = df['Size'].value_counts()

# Plotting the bar chart
plt.subplot(1, 1, 1)
ax = size_distribution.plot(kind='bar', color='lightgreen')

# Adding title and labels
plt.title('Size Distribution')
plt.xlabel('Size')
plt.ylabel('Count')
plt.xticks(rotation=45)

# Adding labels to each bar
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height}', 
                (p.get_x() + p.get_width() / 2., height), 
                ha='center', va='center', 
                xytext=(0, 5), 
                textcoords='offset points')

plt.show()
```

```python
quantity_distribution = df.groupby('Category')['Qty'].sum().sort_values(ascending=False)
plt.subplot(1, 1, 1)
ax = quantity_distribution.plot(kind='bar', color='salmon')
plt.title('Total Quantity Sold by Category')
plt.xlabel('Category')
plt.ylabel('Total Quantity')
plt.xticks(rotation=45)
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height}', 
                (p.get_x() + p.get_width() / 2., height), 
                ha='center', va='center', 
                xytext=(0, 5), 
                textcoords='offset points')
plt.show()
```

# Summary:
Product Distribution: The most popular products are T-shirts and Shirts, with over 49,000 units sold each.


Size Distribution: The most common sizes sold are M and L, followed closely by XL.
Total Quantity Sold: T-shirts and Shirts again lead in total quantities sold, indicating strong demand for these items.

```python
top_names = df['ship-city'].value_counts().nlargest(10)
plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x=top_names.index, y=top_names.values, palette="viridis")
for container in bar_plot.containers:
    bar_plot.bar_label(container)
plt.xlabel('City')
plt.ylabel('Count')
plt.title('Top 10 Ship Cities')
plt.xticks(rotation=45)
plt.show()
```

```python
city_sales = df.groupby('ship-city')['Amount'].sum()

# Filter the sales data for the top 10 cities
top_city_sales = city_sales[top_names.index]

# Create the bar plot
plt.figure(figsize=(12, 6))
top_city_sales.plot(kind='bar', color='skyblue')  # Bar plot with a sky blue color
plt.title('Total Sales in Top 10 Cities by Number of Orders')
plt.xlabel('City')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')  # Rotate x-ticks for better readability
plt.tight_layout()  # Adjust layout to prevent clipping

for index, value in enumerate(top_city_sales):
    plt.text(index, value, f'{value:,.2f}', ha='center', va='bottom')

# Display the plot
plt.show()
```
